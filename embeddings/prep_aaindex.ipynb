```python
df = pd.read_csv("AAidx.txt", sep='\t')
list_amino_acids = df.columns[1:]

total = 0
number = 0
for aa in df.columns[1:-1]:
    total += df[aa]
    number += 1
df['X'] = round(total / number, 2)
df['J'] = round((df['I'] + df['L']) / 2, 2)
df['Z'] = round((df['E'] + df['Q']) / 2, 2)
df['B'] = round((df['D'] + df['N']) / 2, 2)
df
```
