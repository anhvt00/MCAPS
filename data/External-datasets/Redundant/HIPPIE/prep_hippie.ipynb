```python
# Create hippie uni dataframe
df_hip_uni = pd.read_csv('dict_hip_uni.tsv', header=None, sep='\t')
df_hip_uni.columns = ['hip', 'uni']
df_hip_uni
```

```python
# Create hip uni dictionary
dict_hip_uni = {hip: uni for hip, uni in zip(df_hip_uni['hip'].values, df_hip_uni['uni'].values)}
dict_hip_uni
```

```python
def id_to_uni(id):
    if id in dict_hip_uni.keys():
        return dict_hip_uni[id]
    else:
        return 0
```

```python
# Create hip dataframe with uni id
hip_transformed = hip_orig.applymap(lambda key: id_to_uni(key))
hip_transformed['score'] = hip_orig['score']
hip_transformed
```

```python
# Filter all records with non-zero first and second protein
hip_transformed = hip_transformed[(hip_transformed['first'] != 0)  & (hip_transformed['second'] != 0)].reset_index(drop=True)
hip_transformed
```

```python
# Write to pair to create interaction dataset
hip_transformed.to_csv('uni_pairs.tsv', sep='\t', header=None, index=None)
```

```python
# Load uni seq dataframe
uni_seq = pd.read_csv('uni_seq.fasta', sep='\t', names=['uni', 'seq'])
uni_seq
```

```python
# Create uni seq dictionary
dict_uni_seq = {uni: seq for uni, seq in zip(uni_seq['uni'].values, uni_seq['seq'].values)}
dict_uni_seq
```

```python
# Create new dataset HPRD compare to pan
pan_df = pd.read_csv('pan_uni_pairs.tsv', header=None, sep='\t')
pan_df.columns = ['first', 'second', 'label']
pan_df
```

```python
list_pairs_pan = []
for i in range(len(pan_df)):
    list_pairs_pan.append(frozenset(pan_df.iloc[i,:]))
list_pairs_pan
```

```python
list_pairs_hip = []
for i in range(len(hip_transformed)):
    list_pairs_hip.append(frozenset(hip_transformed.iloc[i, :]))
list_pairs_hip
```

```python
# Dictionary for pairs to index
dict_pairs_hip = {key: value for key, value in zip(list_pairs_hip, [x for x in range(len(list_pairs_hip))])}
```

```python
# List of new pairs
list_pairs_hip_new = []
for pairs in list_pairs_hip:
    if pairs not in list_pairs_pan:
        list_pairs_hip_new.append(pairs)
list_pairs_hip_new
```

```python
# ID for new pairs of interactions
id_new = [dict_pairs_hip[key] for key in list_pairs_hip_new]
id_new
```
