```python
import pandas as pd
import numpy as np
```

```python
df_pairs = pd.read_csv('dip_new.tsv', sep='\t', header=None)
df_pairs.columns = ['id1', 'id2', 'label']
df_pairs
```

```python
df_id_seq = pd.read_csv('tempt.txt', header=None, sep='\t')
df_id_seq.columns = ['id', 'seq']
list_protein_cdhit = list(df_id_seq['id'])
list_protein_cdhit
```

```python
def map_id(id):
    if id in list_protein_cdhit:
        return id
    else:
        return 0
```

```python
df_pairs_transformed = df_pairs.applymap(lambda x: map_id(x))
df_pairs_transformed['label'] = df_pairs['label']
df_pairs_transformed
```

```python
df_pairs_transformed = df_pairs_transformed[(df_pairs_transformed['id1'] != 0) & (df_pairs_transformed['id2'] != 0)].reset_index(drop=True)
df_pairs_transformed.to_csv('dip_new_cd40.tsv', header=None, index=None, sep='\t')
```
