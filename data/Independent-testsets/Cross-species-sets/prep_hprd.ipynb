```python
# Import libraries
import pandas as pd
import numpy as np
```

```python
# Original hprd dataset
hprd_orig = pd.read_csv('hprd_2010_pairs.tsv', sep='\t', header=None)
hprd_orig.columns = ['first', 'second', 'label']
hprd_orig
```

```python
# Create set of proteins in hprd
set_hprd_orig = set(list(hprd_orig['first'])+list(hprd_orig['second']))
len(set_hprd_orig), set_hprd_orig
```

```python
# Write set of proteins in hprd original to file
f = open('list.txt', 'w')
for id in set_hprd_orig:
    f.write(id+'\n')
```

```python
# Create ref_uni dataframe
df_ref_uni = pd.read_csv('dict_ref_uni.tsv', header=None, sep='\t')
df_ref_uni.columns = ['ref', 'uni']
df_ref_uni
```

```python
# Create ena_uni dataframe
df_ena_uni = pd.read_csv('dict_ena_uni.tsv', header=None, sep='\t')
df_ena_uni.columns = ['ena', 'uni']
df_ena_uni
```

```python
# Create uni_uni dataframe
df_uni_orig_uni = pd.read_csv('dict_uni_orig_uni.tsv', header=None, sep='\t')
df_uni_orig_uni.columns = ['uni_orig', 'uni']
df_uni_orig_uni
```

```python
# Create ref uni dictionary
dict_ref_uni = {ref: uni for ref, uni in zip(df_ref_uni['ref'].values, df_ref_uni['uni'].values)}
dict_ref_uni
```

```python
# Create ena uni dictionary
dict_ena_uni = {ena: uni for ena, uni in zip(df_ena_uni['ena'].values, df_ena_uni['uni'].values)}
dict_ena_uni
```

```python
# Create uni uni dictionary
dict_uni_orig_uni = {uni_orig: uni for uni_orig, uni in zip(df_uni_orig_uni['uni_orig'].values, df_uni_orig_uni['uni'].values)}
dict_uni_orig_uni
```

```python
def id_to_uni(id):
    if id in dict_ref_uni.keys():
        return dict_ref_uni[id]
    elif id in dict_ena_uni.keys():
        return dict_ena_uni[id]
    elif id in dict_uni_orig_uni.keys():
        return dict_uni_orig_uni[id]
    else:
        return 0
```

```python
# Create pan dataframe with uni id
hprd_transformed = hprd_orig.applymap(lambda key: id_to_uni(key))
hprd_transformed['label'] = hprd_orig['label']
hprd_transformed
```

```python
# Filter all records with non-zero first and second protein
hprd_transformed = hprd_transformed[(hprd_transformed['first'] != 0)  & (hprd_transformed['second'] != 0)].reset_index(drop=True)
hprd_transformed
```

```python
# Write to pair to create interaction dataset
hprd_transformed.to_csv('uni_pairs.tsv', sep='\t', header=None, index=None)
```

```python
# Select human only
all_organism_set = set(list(hprd_transformed['first'])+list(hprd_transformed['second']))
f = open('protein_set_all_organims.txt', 'w')
for entry in all_organism_set:
    f.write(entry+'\n')

# Upload file on uniprot swiss prot id mapping, select interaction filter to check partner and organism to check whether belongs to human
# All is correct!# 
```

```python
# Load uni seq dataframe
uni_seq = pd.read_csv('uni_seq.fasta', sep='\t', names=['uni', 'seq'])
uni_seq
```

```python
# Create uni seq dictionary
dict_uni_seq = {uni: seq for uni, seq in zip(uni_seq['uni'].values, uni_seq['seq'].values)}
dict_uni_seq
```

```python
# Create new dataset HPRD compare to pan
pan_df = pd.read_csv('pan_uni_pairs.tsv', header=None, sep='\t')
pan_df.columns = ['first', 'second', 'label']
pan_df
```

```python
list_pairs_pan = []
for i in range(len(pan_df)):
    list_pairs_pan.append(frozenset(pan_df.iloc[i,:]))
list_pairs_pan
```

```python
list_pairs_hprd = []
for i in range(len(hprd_transformed)):
    list_pairs_hprd.append(frozenset(hprd_transformed.iloc[i, :]))
list_pairs_hprd
```

```python
# Dictionary for pairs to index
dict_pairs_hprd = {key: value for key, value in zip(list_pairs_hprd, [x for x in range(len(list_pairs_hprd))])}
```

```python
# List of new pairs
list_pairs_hprd_new = []
for pairs in list_pairs_hprd:
    if pairs not in list_pairs_pan:
        list_pairs_hprd_new.append(pairs)
list_pairs_hprd_new
```

```python
# ID for new pairs of interactions
id_new = [dict_pairs_hprd[key] for key in list_pairs_hprd_new]
id_new
```

```python
# Write to file all new interaction pairs in dataset
hprd_new = hprd_transformed.iloc[id_new, :]
hprd_new.to_csv('hprd_new.tsv', sep='\t', header=None, index=None)
hprd_new
```

```python
# Write to file list of new protein
set_protein_new = list(hprd_new['first'])+list(hprd_new['second'])
dict_protein_new = {id: seq for id, seq in zip(set_protein_new, [dict_uni_seq[id] for id in set_protein_new])}
```
